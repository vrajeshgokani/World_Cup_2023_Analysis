### Team Performance Analysis

```python
df1 = df.sort_values(by=["team","bat_or_bowl"])
df1
```

```python
team1 = df1.groupby(['team','bat_or_bowl']).agg({
    'runs': 'sum'}).reset_index()
team1
```

Top Batting Teams(Most number of runs scored)

```python
team1 = team1[team1['bat_or_bowl'] == 'bat']
top_batting_teams = team1.sort_values(by=["runs"],ascending=False)
top_batting_teams
```

```python
team2 = df1.groupby('team').agg({
    'wkts': 'sum',}).reset_index()
team2
```

Top Bowling Teams(Most Wickets taken)

```python
top_bowling_teams = team2.sort_values(by=["wkts"],ascending=False)
top_bowling_teams
```

```python
teamoverall= df1.groupby('team').agg({
    'wkts': 'sum',
    'runs': 'sum'}).reset_index()
teamoverall
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12, 8))
bar_width = 0.4
bar_positions_runs = range(len(teamoverall['team']))
ax1.bar(bar_positions_runs, team1['runs'], width=bar_width, label='Runs', color='b')
ax1.set_xlabel('Team')
ax1.set_ylabel('Runs', color='b')
ax1.set_xticks([pos + bar_width / 2 for pos in bar_positions_runs])
ax1.set_xticklabels(teamoverall['team'], rotation=45, ha='right')
ax1.tick_params(axis='y', labelcolor='b')
ax1.legend(loc='upper left')
ax2 = ax1.twinx()
color = 'r'
ax2.set_ylabel('Wickets', color=color)
ax2.bar([pos + bar_width for pos in bar_positions_runs], teamoverall['wkts'], width=bar_width, label='Wickets', color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')
for ax in [ax1, ax2]:
    for bar in ax.patches:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')
plt.title('Team Performance')
plt.show()
```

## Player Performance Analysis

```python
df
df2 = df.sort_values(by=["player"])
df2
```

```python
df2 = df2[df2['bat_or_bowl'] == 'bat']
batsman = df2.groupby('player').agg({
    'runs': 'sum',
    'bb_bf': 'sum',
    'sr':'mean',
    '4s':'sum',
    '6s':'sum'}).reset_index()

top_scorers = batsman.sort_values(by=["runs"],ascending=False)
top_scorers.head(10)
```

```python
import matplotlib.pyplot as plt

top_10_batsmen = top_scorers.head(10)

plt.figure(figsize=(12, 8))
bars = plt.bar(top_10_batsmen['player'], top_10_batsmen['runs'], color='yellow', edgecolor='black', linewidth=1.2)

for bar, value in zip(bars, top_10_batsmen['runs']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 2, f"{round(value, 2)}",
             ha='center', va='bottom', color='black', fontweight='bold', fontsize=10)

plt.xlabel('Players', fontsize=14, fontweight='bold')
plt.ylabel('Runs Scored', fontsize=14, fontweight='bold')
plt.title('Most Runs Scored', fontsize=16, fontweight='bold')
plt.ylim(0, max(top_10_batsmen['runs']) + 100)
plt.xticks(rotation=45, ha='right', fontsize=12)

plt.grid(axis='y', linestyle='-', alpha=0.7)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.gca().set_facecolor('#F5F5F5')

plt.tight_layout()
plt.show()
```

```python
batsman = batsman[batsman['bb_bf'] >= 100]
best_strikerate = batsman.sort_values(by=["sr"],ascending=False)
best_strikerate.head(10)
```

## Opposition And Ground Analysis

```python
import pandas as pd
df=pd.read_csv("/content/CWC23_all_innings.csv")
```

```python
teamopp = df[df['bat_or_bowl']=='bat']
teamopp = teamopp.groupby(['team', 'opposition']).agg({
    'runs': 'sum'
}).reset_index()
condition = (teamopp['team'] == 'NED')&(teamopp['opposition'] == 'v Netherlands')

teamopp = teamopp.drop(teamopp[condition].index)
teamopp
```

```python
teamopp1 = df[df['bat_or_bowl']=='bowl']
teamopp1 = teamopp1.groupby(['team', 'opposition']).agg({
    'wkts': 'sum'
}).reset_index()
condition = (teamopp1['team'] == 'NED')&(teamopp1['opposition'] == 'v Netherlands')

teamopp1 = teamopp1.drop(teamopp1[condition].index)
teamopp1
```

```python
import matplotlib.pyplot as plt
import numpy as np

teams = teamopp['team'].unique()

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))

for i, team in enumerate(teams):
    row = i // 2
    col = i % 2

    team_data = teamopp[teamopp['team'] == team]

    x_positions = np.arange(len(team_data['opposition']))

    axes[row, col].plot(x_positions, team_data['runs'], marker='o', label='Runs', color='#1f77b4', alpha=0.7, linestyle='-', linewidth=2)

    ax2 = axes[row, col].twinx()

    team_data = teamopp1[teamopp1['team'] == team]
    ax2.plot(x_positions, team_data['wkts'], marker='s', label='Wickets', color='#2ca02c', alpha=0.7, linestyle='-', linewidth=2)

    axes[row, col].set_xticks(x_positions)
    axes[row, col].set_xticklabels(team_data['opposition'], rotation=45, ha='right')

    axes[row, col].set_ylabel(f'{team} Runs', color='#1f77b4')
    ax2.set_ylabel(f'{team} Wickets', color='#2ca02c')

    axes[row, col].legend(loc='upper left', frameon=False)
    ax2.legend(loc='upper right', frameon=False)

    axes[row, col].grid(axis='y', linestyle='-', alpha=0.7)

    axes[row, col].set_title(f'{team} Performance')

plt.xlabel('Opposition')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from tabulate import tabulate

batsmen_data = df[df['bat_or_bowl'] == 'bat']

batting_team_performance = batsmen_data.groupby(['team', 'opposition', 'ground'])['runs'].sum().reset_index()

best_batting_team = batting_team_performance.loc[batting_team_performance.groupby('ground')['runs'].idxmax()]

print("Best Batting Team on Each Ground:")
print(tabulate(best_batting_team[['ground', 'team', 'runs', 'opposition']],
               headers=['Ground', 'Best Batting Team', 'Runs', 'Opposition'], tablefmt='pretty'))

bowlers_data = df[df['bat_or_bowl'] == 'bowl']

bowling_team_performance = bowlers_data.groupby(['team', 'opposition', 'ground'])['wkts'].sum().reset_index()

best_bowling_team = bowling_team_performance.loc[bowling_team_performance.groupby('ground')['wkts'].idxmax()]

print("\nBest Bowling Team on Each Ground:")
print(tabulate(best_bowling_team[['ground', 'team', 'wkts', 'opposition']],
               headers=['Ground', 'Best Bowling Team', 'Wickets', 'Opposition'], tablefmt='pretty'))
```

## Temporal Analysis

```python
import matplotlib.pyplot as plt

team_runs_wickets_over_time = df.groupby(['team', 'start_date'])[['runs', 'wkts']].sum().reset_index()

unique_teams = team_runs_wickets_over_time['team'].unique()
num_teams = len(unique_teams)

fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 15))
axes = axes.flatten()

for i, team in enumerate(unique_teams):
    team_data = team_runs_wickets_over_time[team_runs_wickets_over_time['team'] == team]

    ax1 = axes[i]
    color = 'tab:red'
    ax1.set_xlabel('Start Date')
    ax1.set_ylabel('Runs', color=color)
    ax1.plot(team_data['start_date'], team_data['runs'], label='Runs', marker='o', linestyle='-', color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    ax1.legend(loc='upper left')
    plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha='right')

    ax2 = ax1.twinx()
    color = 'tab:blue'
    ax2.set_ylabel('Wickets', color=color)
    ax2.plot(team_data['start_date'], team_data['wkts'], label='Wickets', marker='s', linestyle='--', color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    ax2.legend(loc='upper right')
    plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha='right')

    ax1.grid(True, linestyle='--', alpha=0.7)
    ax2.grid(True, linestyle='--', alpha=0.7)

    plt.title(f'{team} Runs and Wickets Over Time')

plt.tight_layout()
plt.show()
```
